```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load Iris dataset
iris = sns.load_dataset("iris")
print(iris.head())

chosen_vars = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

plt.figure(figsize=(8,5))
sns.boxplot(data=iris[chosen_vars], orient="h")
plt.title("Boxplots of selected Iris features")
plt.show()
```

```python
# Basic stats: mean, std, min, max, quartiles
print(iris.describe())
plt.figure(figsize=(8,5))
sns.boxplot(data=iris[chosen_vars],orient="h")
plt.title("Boxplots of selected Iris fatures")
```

```python
print("Variance:\n", iris.var(numeric_only=True))
print("\nStandard Deviation:\n", iris.std(numeric_only=True))
# Basic stats: mean, std, min, max, quartiles
print(iris.describe())
print("Variance:\n", iris.var(numeric_only=True))
print("\nStandard Deviation:\n", iris.std(numeric_only=True))
iris.hist(figsize=(10,6), bins=20)
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Width by Species")
plt.show()
```
